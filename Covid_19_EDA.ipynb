```python
# Total Data Analysis for different columns

Total_Cases = df['new_cases'].sum()
print("Total Cases:",Total_Cases)

Total_Deaths = df['new_deaths'].sum()
print("Total Deaths:", Total_Deaths)

Total_Recovered = Total_Cases - Total_Deaths
print("Total Recovered:", Total_Recovered)

Total_Tests = df['new_tests'].sum()
print("Total Tests:", Total_Tests)

Total_Vaccinations = df['new_vaccinations'].sum()
print("Total Vaccinations:", Total_Vaccinations/2)

Avg_Life_Exp = df['life_expectancy'].mean()
print("Average Life Expectancy", Avg_Life_Exp,'%')
```

# Data Visualization

```python
# Continent Wise Analysis

# 1. Total Cases Per Continent

covid = df.groupby('continent')['new_cases'].sum()

x = covid.index
y = covid

plt.figure(figsize = (8,4))
sns.barplot(x = x, y = y, hue = x ,palette = 'bright')
plt.xlabel("Continents", size = 12)
plt.ylabel("Total Cases(in billions)", size = 12)
plt.title("Total Covid Cases Per Continent")
plt.show()
```

```python
# 2. Total Deaths Per Continent

death = df.groupby('continent')['new_deaths'].sum()

x = death.index
y = death

plt.figure(figsize = (8,4))
sns.barplot(x = x, y = y, hue = x ,palette = 'Paired')
plt.xlabel("Continents", size = 12)
plt.ylabel("Total Deaths(in millions)", size = 12)
plt.title("Total Deaths Cases Per Continent")
plt.show()
```

```python
# 3. Total Recovered Cases Per Continent

recovered = df.groupby('continent')['recovered'].sum()

x = recovered.index
y = recovered

plt.figure(figsize = (8,4))
sns.barplot(x = x, y = y, hue = x ,palette = 'bright')
plt.xlabel("Continents", size = 12)
plt.ylabel("Total Recovered(in billions)", size = 12)
plt.title("Total Recovered Cases Per Continent")
plt.show()
```

```python
# 4. Average Patients Age Per Continent

median = df.groupby('continent')['median_age'].mean()

x = median.index
y = median

plt.figure(figsize = (8,4))
sns.barplot(x = x, y = y, hue = x ,palette = 'Paired')
plt.xlabel("Continents", size = 12)
plt.ylabel("Average Age for Patients(in years)", size = 12)
plt.title("Average Age Per Continent")
plt.show()
```

```python
# 5. Average Life Expectancy Per Continent

life = df.groupby('continent')['life_expectancy'].mean()

x = life.index
y = life

plt.figure(figsize = (8,4))
sns.barplot(x = x, y = y, hue = x ,palette = 'bright')
plt.xlabel("Continents", size = 12)
plt.ylabel("Average Life Expectancy(in years)", size = 12)
plt.title("Average Life Expectancy Per Continent")
plt.show()
```

```python
# 6. Total Cardio Vascular Deaths Per Continent

cardio = df.groupby('continent')['cardiovasc_death_rate'].sum()

x = cardio.index
y = cardio

plt.figure(figsize = (8,4))
sns.barplot(x = x, y = y, hue = x ,palette = 'Paired')
plt.xlabel("Continents", size = 12)
plt.ylabel("Cardio Vascular Deaths for Patients", size = 12)
plt.title("Cardio Vascular Deaths Per Continent")
plt.show()
```
